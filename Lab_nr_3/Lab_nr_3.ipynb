```python
import pandas as pd
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
data = pd.read_csv('reviews_courses.csv', parse_dates=['Timestamp'])
print(f'Nasza dataset zawiera: \n- {data.shape[0]} wierszy, oraz {data.shape[1]} kolumny')
# Histogram z rozkładem wartości dla kolumny <Ratings>
# data['Rating'].hist()
# Wyświetlanmy wartość dla integer w kolumnie po indeksu
data.iloc[3]['Rating']
average_value = data['Rating'].mean()
print(f"Średnia dla wszystkich kursów wynosi: {average_value}")
Rating = data[data['Rating']  >  4]['Rating'].mean() 
print(f"Średnia dla wszystkich kursów większych cztery wynosi: {Rating}")
# Funkcja count() - wyświetla ilość która jest w kolumnach nie ważny typ danych
ilosc = data[data['Course Name'] == 'Learn GIS in One Hour']['Rating'].count()
print(f"Ilość wynosi: {ilosc}")      
zad_3_2 = data[
    (data['Rating'] > 4) & 
    (data['Course Name'] == 'The Python Course: OpenCV in Python')
]

zad_4 = data[
    (data['Timestamp'] >= datetime(2020, 7, 1, tzinfo = utc)) &
    (data['Timestamp'] >= datetime(2020, 12, 31, tzinfo = utc))
]

zad_5_1 = data[
    (data['Course Name'] == 'The Python Course: From Beginner to Expert') &
    (data['Timestamp'] >= datetime(2020, 7, 1, tzinfo = utc)) &
    (data['Timestamp'] >= datetime(2020, 12, 31, tzinfo = utc))
]['Rating'].mean()
print(f'Średnia ocen dla \n- kursu "The Python Course: From Beginner to Expert",\n- w tym, jedynie od 1 lipca 2020 do 31 grudnia 2020 wynosi: {zad_5_1}')
data['Day'] = data['Timestamp'].dt.date
day_avg = data.groupby(['Day']).mean()
```
