```python
import pandas as pd
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
data = pd.read_csv('reviews_courses.csv', parse_dates=['Timestamp'])

# Wykres według dni
data['Day'] = data['Timestamp'].dt.date
data_day = data.groupby(['Day']).mean()
# plt.figure(figsize=(20, 8))
# plt.plot(data_day.index, data_day['Rating'])

# Wykres według roku i tygodnia
data['Week'] = data['Timestamp'].dt.strftime('%Y-%U')
data_week = data.groupby(['Week']).mean()
# plt.figure(figsize=(20, 8))
# plt.plot(data_week.index, data_week['Rating'])

# Wykres według roku i miesiąca
data['Month'] = data['Timestamp'].dt.strftime('%Y-%m')
data_month = data.groupby(['Month']).mean()
# plt.figure(figsize=(20, 8))
# plt.plot(data_month.index, data_month['Rating'])

# Wykres według dni tygodnia
data['Weekday'] = data['Timestamp'].dt.strftime('%A')
# Dodanie numeru dnia tygodnia
data['NumberDay'] = data['Timestamp'].dt.strftime('%w')
# Grupujemy dane
data_weekday = data.groupby(['Weekday','NumberDay']).mean()
# sortujemy według wartości numerycznych
data_weekday_number_day = data_weekday.sort_values('NumberDay')

# Wyświetlamy wykres
plt.figure(figsize=(20, 10))
plt.plot(
    data_weekday_number_day.index.get_level_values(0), 
    data_weekday_number_day['Rating']
)


# data_month_course = data.groupby(['Month', 'Course Name']).mean().unstack()
# data_month_course.plot(figsize=(20, 8))

# data_month_course = data.groupby(['Month', 'Course Name'])['Rating'].count().unstack()
# data_month_course.plot(figsize=(20, 20))


```
